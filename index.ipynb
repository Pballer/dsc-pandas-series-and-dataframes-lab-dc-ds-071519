## Group the Data by Day of Week and Plot the Sum of The Numeric Columns

```python
df.Date = pd.to_datetime(df.Date, format='%m/%d/%Y')
```

```python
#Your code here
day_sums = df.groupby(df.Date.dt.dayofweek).sum()
```
